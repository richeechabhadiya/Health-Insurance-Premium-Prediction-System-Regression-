```python
# ===============================
# 03_data_preprocessing.ipynb
# ===============================

# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Step 2: Load cleaned dataset from feature analysis
df = pd.read_csv("/home/richee/Downloads/Assesment/data/processed/clean_data.csv")
print("Cleaned dataset loaded. Shape:", df.shape)

# Step 3: Separate features and target
target_col = "premium"
X = df.drop(columns=[target_col])
y = df[target_col]

print("\nFeatures shape:", X.shape)
print("Target shape:", y.shape)

# Step 4: Identify numeric & categorical columns
numeric_cols = X.select_dtypes(include=np.number).columns.tolist()
categorical_cols = X.select_dtypes(exclude=np.number).columns.tolist()

print("\nNumeric columns:", numeric_cols)
print("Categorical columns:", categorical_cols)

# Step 5: Handle categorical encoding
# Updated for scikit-learn >= 1.2
ohe = OneHotEncoder(sparse_output=False, drop='first')  # drop first to avoid dummy trap
X_encoded = pd.DataFrame(
    ohe.fit_transform(X[categorical_cols]),
    columns=ohe.get_feature_names_out(categorical_cols)
)

# Step 6: Combine numeric and encoded categorical features
X_final = pd.concat([X[numeric_cols].reset_index(drop=True), X_encoded.reset_index(drop=True)], axis=1)

print("\nFinal feature matrix shape:", X_final.shape)

# Optional: Scaling numeric features (good for Linear Regression, Ridge, etc.)
scaler = StandardScaler()
X_final[numeric_cols] = scaler.fit_transform(X_final[numeric_cols])
print("Numeric features scaled (mean=0, std=1)")

# Step 7: Save processed features and target
import os

# Ensure folder exists
os.makedirs("../data/processed", exist_ok=True)

# Save files
X_final.to_csv("../data/processed/X_final.csv", index=False)
y.to_csv("../data/processed/y_final.csv", index=False)
print("Processed files saved successfully!")
```
