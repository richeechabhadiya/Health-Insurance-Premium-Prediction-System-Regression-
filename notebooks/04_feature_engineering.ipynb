```python
# ===============================
# 04_feature_engineering.ipynb
# ===============================

# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import os

print("Step 1: Libraries imported.")

# Step 2: Load preprocessed dataset
X_final = pd.read_csv("../data/processed/X_final.csv")
y_final = pd.read_csv("../data/processed/y_final.csv")

print(f"Step 2: Preprocessed dataset loaded. X shape: {X_final.shape}, y shape: {y_final.shape}")

# Step 3: Create business-driven derived features
# -----------------------------------------------
# Age group (if not present)
if 'age_group' not in X_final.columns and 'age' in X_final.columns:
    X_final['age_group'] = pd.cut(
        X_final['age'],
        bins=[0,25,35,45,55,65,75,100],
        labels=['18-25','26-35','36-45','46-55','56-65','66-75','76+']
    )
    print("Step 3a: Created age_group feature.")

# Seniority difference
if 'seniority_diff' not in X_final.columns and 'seniority_policy' in X_final.columns and 'seniority_insured' in X_final.columns:
    X_final['seniority_diff'] = X_final['seniority_policy'] - X_final['seniority_insured']
    print("Step 3b: Created seniority_diff feature.")

# Claims per medical service
if 'claims_per_service' not in X_final.columns and 'cost_claims_year' in X_final.columns and 'n_medical_services' in X_final.columns:
    X_final['claims_per_service'] = X_final['cost_claims_year'] / (X_final['n_medical_services'] + 1)
    print("Step 3c: Created claims_per_service feature.")

# Premium per exposure year
if 'premium_per_year' not in X_final.columns and 'premium' in y_final.columns and 'exposure_time' in X_final.columns:
    X_final['premium_per_year'] = y_final['premium'] / (X_final['exposure_time'] + 0.1)
    print("Step 3d: Created premium_per_year feature.")

# Region risk index
if 'region_risk' not in X_final.columns and 'IICIMUN' in X_final.columns and 'IICIPROV' in X_final.columns:
    X_final['region_risk'] = X_final['IICIMUN'] * 0.6 + X_final['IICIPROV'] * 0.4
    print("Step 3e: Created region_risk feature.")

# Step 4: Interaction features
# -----------------------------
if 'age' in X_final.columns and 'n_medical_services' in X_final.columns:
    X_final['age_medical_interaction'] = X_final['age'] * X_final['n_medical_services']

if 'seniority_insured' in X_final.columns and 'cost_claims_year' in X_final.columns:
    X_final['seniority_claims_interaction'] = X_final['seniority_insured'] * X_final['cost_claims_year']

# Optional: channel × product interaction
if 'distribution_channel_A' in X_final.columns and 'type_product_S' in X_final.columns:
    X_final['channel_product_interaction'] = X_final['distribution_channel_A'] * X_final['type_product_S']

print("Step 4: Interaction features created.")

# Step 5: Scale numerical features (optional for linear models)
num_features = [col for col in X_final.select_dtypes(include=np.number).columns.tolist()]
scaler = StandardScaler()
X_final[num_features] = scaler.fit_transform(X_final[num_features])
print("Step 5: Numeric features scaled (mean=0, std=1)")

# Step 6: Save the enhanced feature set
os.makedirs("../data/processed", exist_ok=True)
X_final.to_csv("../data/processed/X_final.csv", index=False)
y_final.to_csv("../data/processed/y_final.csv", index=False)
print("Step 6: Feature engineering complete. Files saved to data/processed/")
print(f"Final X shape: {X_final.shape}, y shape: {y_final.shape}")
```

```python
# ===============================
# 04_feature_engineering.ipynb  (FIXED - no leakage)
# ===============================

# Step 1: Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import os

print('Step 1: Libraries imported.')


# -----------------------------------------------
# Step 2: Load preprocessed dataset
# -----------------------------------------------
# Load from clean_data (NOT from X_final — we rebuild X_final cleanly here)
df = pd.read_csv('../data/processed/clean_data.csv')
print(f'Step 2: clean_data loaded. Shape: {df.shape}')


# -----------------------------------------------
# Step 3: Separate target FIRST before anything
# -----------------------------------------------
target_col = 'premium'
y_final = df[[target_col]].copy()
X_final = df.drop(columns=[target_col]).copy()

print(f'Target (y) shape: {y_final.shape}')
print(f'Features (X) shape: {X_final.shape}')


# -----------------------------------------------
# Step 4: DROP LEAKAGE COLUMNS
# These columns either ARE the target, come from
# the future, or are outcomes — not inputs.
# -----------------------------------------------
leakage_cols = [
    # IDs — not predictive
    'ID', 'ID_policy', 'ID_insured',

    # Future/lapse info — not known at policy start
    'year_lapse_insured', 'year_lapse_policy',
    'date_lapse_insured', 'date_lapse_policy',
    'lapse',

    # ⚠️ CLAIM COLUMNS — these are OUTCOMES of the policy,
    # not inputs available when pricing the premium.
    # Keeping them = model memorises answers, not patterns.
    'cost_claims_year',
    'n_medical_services',
]

X_final = X_final.drop(columns=leakage_cols, errors='ignore')
print(f'Step 4: After dropping leakage columns. Shape: {X_final.shape}')
print(f'Remaining columns: {list(X_final.columns)}')


# -----------------------------------------------
# Step 5: LEGITIMATE Feature Engineering
# (Only using columns available at policy-start time)
# -----------------------------------------------

# Age group buckets
if 'age' in X_final.columns:
    X_final['age_group'] = pd.cut(
        X_final['age'],
        bins=[0, 25, 35, 45, 55, 65, 75, 100],
        labels=['18-25', '26-35', '36-45', '46-55', '56-65', '66-75', '76+']
    ).astype(str)
    print('Step 5a: Created age_group.')

# Seniority difference (policy vs insured)
if 'seniority_policy' in X_final.columns and 'seniority_insured' in X_final.columns:
    X_final['seniority_diff'] = X_final['seniority_policy'] - X_final['seniority_insured']
    print('Step 5b: Created seniority_diff.')

# Region risk index (from geographic cols — no claim data)
if 'IICIMUN' in X_final.columns and 'IICIPROV' in X_final.columns:
    iicimun = pd.to_numeric(X_final['IICIMUN'], errors='coerce').fillna(0)
    iiciprov = pd.to_numeric(X_final['IICIPROV'], errors='coerce').fillna(0)
    X_final['region_risk'] = iicimun * 0.6 + iiciprov * 0.4
    print('Step 5c: Created region_risk.')

# Age interactions with seniority (both known at policy start)
if 'age' in X_final.columns and 'seniority_insured' in X_final.columns:
    X_final['age_x_seniority'] = X_final['age'] * X_final['seniority_insured']
    print('Step 5d: Created age_x_seniority.')

# Age interactions with exposure (both known at policy start)
if 'age' in X_final.columns and 'exposure_time' in X_final.columns:
    X_final['age_x_exposure'] = X_final['age'] * X_final['exposure_time']
    print('Step 5e: Created age_x_exposure.')

# Non-linear age (insurance pricing is non-linear with age)
if 'age' in X_final.columns:
    X_final['age_squared'] = X_final['age'] ** 2
    print('Step 5f: Created age_squared.')

# Geographic density ratios
if 'n_insured_prov' in X_final.columns and 'n_insured_mun' in X_final.columns:
    X_final['prov_mun_ratio'] = (
        pd.to_numeric(X_final['n_insured_prov'], errors='coerce') /
        (pd.to_numeric(X_final['n_insured_mun'], errors='coerce') + 1)
    )
    print('Step 5g: Created prov_mun_ratio.')

# Distribution channel x product interaction
if 'distribution_channel_A' in X_final.columns and 'type_product_S' in X_final.columns:
    X_final['channel_product_interaction'] = (
        X_final['distribution_channel_A'] * X_final['type_product_S']
    )
    print('Step 5h: Created channel_product_interaction.')

print(f'\nStep 5: Feature engineering done. Shape: {X_final.shape}')


# -----------------------------------------------
# Step 6: Handle Missing Values
# -----------------------------------------------
# Numeric columns → fill with median
num_cols = X_final.select_dtypes(include=np.number).columns
X_final[num_cols] = X_final[num_cols].apply(lambda c: c.fillna(c.median()))

# Categorical columns → fill with 'Unknown'
cat_cols = X_final.select_dtypes(exclude=np.number).columns
X_final[cat_cols] = X_final[cat_cols].fillna('Unknown')

print('Step 6: Missing values filled.')


# -----------------------------------------------
# Step 7: Encode Categorical Columns
# -----------------------------------------------
# Cap high-cardinality columns to top 10 values before encoding
cat_cols = X_final.select_dtypes(include='object').columns
for col in cat_cols:
    if X_final[col].nunique() > 10:
        top_10 = X_final[col].value_counts().nlargest(10).index
        X_final[col] = np.where(X_final[col].isin(top_10), X_final[col], 'Other')

X_final = pd.get_dummies(X_final, drop_first=True)
print(f'Step 7: After encoding. Shape: {X_final.shape}')


# -----------------------------------------------
# Step 8: Scale Numeric Features
# (StandardScaler — good for linear models)
# -----------------------------------------------
num_features = X_final.select_dtypes(include=np.number).columns.tolist()
scaler = StandardScaler()
X_final[num_features] = scaler.fit_transform(X_final[num_features])
print('Step 8: Numeric features scaled (mean=0, std=1).')


# -----------------------------------------------
# Step 9: Save
# -----------------------------------------------
os.makedirs('../data/processed', exist_ok=True)
X_final.to_csv('../data/processed/X_final.csv', index=False)
y_final.to_csv('../data/processed/y_final.csv', index=False)

print('\nStep 9: Files saved!')
print(f'Final X shape: {X_final.shape}')
print(f'Final y shape: {y_final.shape}')
print(f'Columns in X_final: {list(X_final.columns)}')


# -----------------------------------------------
# Step 10: Sanity Check — make sure NO claim/premium
# columns survived into X_final
# -----------------------------------------------
banned = ['cost_claims', 'n_medical', 'premium', 'claim', 'lapse']
leaked = [c for c in X_final.columns if any(b in c.lower() for b in banned)]
if leaked:
    print(f'\n⚠️  WARNING: Possible leakage columns still in X_final: {leaked}')
    print('   Remove them before training!')
else:
    print('\n✅ Sanity check passed — no leakage columns detected in X_final.')
```
