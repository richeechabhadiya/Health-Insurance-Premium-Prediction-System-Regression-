```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

df = pd.read_excel("/home/richee/Downloads/Assesment/data/raw/Dataset of health insurance portfolio.xls")

df.head()
```

```python
print("\nTarget Variable: premium")

print("Mean:", df["premium"].mean())
print("Median:", df["premium"].median())
print("Standard Deviation:", df["premium"].std())
print("Minimum:", df["premium"].min())
print("Maximum:", df["premium"].max())

print("\nDescription: This helps understand the scale and spread of the target variable.")
```

```python
corr = df.corr(numeric_only=True)

corr_target = corr["premium"].sort_values(ascending=False)

print("\nCorrelation of features with premium:")
display(corr_target)

print("\nDescription: Features with higher absolute correlation values are more important for prediction.")
```

```python
strong_features = corr_target[abs(corr_target) > 0.2].index.tolist()

print("\nSelected strong features (correlation > 0.2):")
print(strong_features)

print("\nDescription: These features have meaningful relationship with the target variable.")
```

```python
corr_matrix = df[strong_features].corr().abs()

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.85)]

print("\nHighly correlated features to remove:")
print(to_drop)

print("\nDescription: Features highly correlated with each other may cause multicollinearity issues.")
```

```python
variance = df.var(numeric_only=True).sort_values(ascending=False)

print("\nFeature variance:")
display(variance)

print("\nDescription: Low variance features contain less information and may be removed.")
```

```python
cat_cols = df.select_dtypes(exclude=np.number).columns

for col in cat_cols:
    print(f"\nAverage premium by {col}:")
    display(df.groupby(col)["premium"].mean().sort_values(ascending=False))

print("\nDescription: This shows how categorical variables affect premium.")
```

```python
final_features = [col for col in strong_features if col not in to_drop]

print("\nFinal selected features:")
print(final_features)

print("\nDescription: These features will be used to train the model.")
```

```python
drop_cols = [
    "ID", "ID_policy", "ID_insured",
    "date_effect_insured", "date_lapse_insured",
    "date_effect_policy", "date_lapse_policy"
]

df = df.drop(columns=drop_cols, errors="ignore")

print("Dropped unnecessary columns (IDs and dates)")
print("Description: These columns do not contribute to prediction.")
```

```python
final_features = [
    # Numerical Features
    "age",
    "exposure_time",
    "seniority_insured",
    "seniority_policy",
    "cost_claims_year",
    "n_medical_services",
    "n_insured_pc",
    "n_insured_mun",
    "n_insured_prov",
    "IICIMUN",
    "IICIPROV",

    # Categorical Features
    "gender",
    "type_policy",
    "type_policy_dg",
    "type_product",
    "reimbursement",
    "new_business",
    "distribution_channel",

    # Target (keep separately)
    "premium"
]

print("\nFinal selected features:")
print(final_features)

print("\nDescription: Features selected using business understanding, EDA insights, and domain relevance.")
```

```python
missing_cols = [col for col in final_features if col not in df.columns]

print("Missing columns:", missing_cols)

print("\nDescription: Ensures all selected features are present in dataset.")
```

```python
null_check = df[final_features].isnull().sum()

print("\nNull values in selected features:")
print(null_check[null_check > 0])

print("\nDescription: Ensures no missing values before modeling.")
```

```python
df["IICIMUN_missing"] = df["IICIMUN"].isnull().astype(int)
df["IICIPROV_missing"] = df["IICIPROV"].isnull().astype(int)

df["IICIMUN"] = df["IICIMUN"].fillna(df["IICIMUN"].median())
df["IICIPROV"] = df["IICIPROV"].fillna(df["IICIPROV"].median())

print("Handled missing values with indicator variables")
```
