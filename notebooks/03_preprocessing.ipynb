```python
# ===============================
# 03_data_preprocessing.ipynb
# ===============================

# Step 1: Import libraries
import pandas as pd
import numpy as np
import os

print("Step 1: Libraries imported.")


# -----------------------------------
# Step 2: Load cleaned dataset
# -----------------------------------
df = pd.read_csv("/home/richee/Downloads/Assesment/data/processed/clean_data.csv")
print("Step 2: Dataset loaded. Shape:", df.shape)


# -----------------------------------
# Step 3: Basic sanity check
# -----------------------------------
print("\nColumns in dataset:")
print(df.columns.tolist())


# -----------------------------------
# Step 4: Handle missing values (ONLY CLEANING)
# -----------------------------------

# Numeric → median
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.median()))

# Categorical → "Unknown"
cat_cols = df.select_dtypes(exclude=np.number).columns
df[cat_cols] = df[cat_cols].fillna("Unknown")

print("\nStep 4: Missing values handled.")


# -----------------------------------
# Step 5: Basic type fixin
# -----------------------------------

# Convert numeric-like columns safely
for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert categorical properly
for col in cat_cols:
    df[col] = df[col].astype(str)

print("Step 5: Data types fixed.")


# -----------------------------------
# Step 6: REMOVE obvious useless columns
# -----------------------------------
drop_cols = [
    "ID", "ID_policy", "ID_insured"
]

df = df.drop(columns=drop_cols, errors='ignore')

print("Step 6: Dropped unnecessary columns.")


# -----------------------------------
# Step 7: Save CLEAN dataset ONLY
# -----------------------------------
os.makedirs("../data/processed", exist_ok=True)

df.to_csv("../data/processed/clean_data_final.csv", index=False)

print("\nClean dataset saved as clean_data_final.csv")
print("Shape:", df.shape)
```
